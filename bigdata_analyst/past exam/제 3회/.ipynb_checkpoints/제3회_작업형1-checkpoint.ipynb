# 문제 2 --------------------------------------------------------------

2000년 데이터 중 2000년 평균보다 큰 값의 데이터 수를 구하시오 (index = 연도
)

```python
import pandas as pd
df = pd.read_csv('year.csv', index_col='Unnamed: 0')
```

```python
df.head()
```

```python
df.mean(axis=0)
```

# [?] index _ col은 왜 준걸까? -> 2000, 과 같은 index의 이름 불러오는 것

```python
df_2000_mean = df.loc[2000,].mean(axis=1)
```

```python
cond1 = df.loc[2000,] > df_2000_mean #행이고 열이고
```

```python
cond1
```

```python
len(df.loc[2000,][cond1])
```
