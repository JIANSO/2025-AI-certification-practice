# 문제 1 --------------------------------------------------------------------

'age'컬럼의 3사분위수와 1사분위수의 차이를 절댓값으로 구하고, 
소수점 이하를 버리고, 정수로 출력하시오.

```python
import pandas as pd
df = pd.read_csv('data4-1.csv')
```

```python
df.head()
```

```python
# 3사분위수와 1사분위수
q3 = df['age'].quantile(0.75)
q1 = df['age'].quantile(0.25)
```

```python
#차이의 절댓값
result = abs(q3-q1)
```

```python
print(result)
```

```python
result = round(result)
print("답 :", result)
```

# 문제 2 --------------------------------------------------------------------

모든 reactions 중 loves와 wows를 합한 반응 비율이 40% 보다 크고 50%보다 작은 데이터를 찾고,
이 중 type 컬럼이 'video'인 데이터의 개수를 구하시오.

```python
import pandas as pd
df = pd.read_csv("data4-2.csv")
```

```python
df.head()
```

```python
# 모든 리액션 (reactions), loves와  wows를 합한 리액션
```

```python
goal  = (df['loves'] + df['wows'])/df['reactions']
```

```python
result = ((goal > 0.4) & (goal < 0.5))
```

```python
df[result]
```

```python
cond2 = df[result]['type'] =='video'
```

```python
cond2
```

```python
len(df[result][cond2])
```

# 문제 3 --------------------------------------------------------------------

'date_added' 컬럼의 날짜가 2018년 1월이면서 'country'에서 United Kingdom 단독 제작인 데이터의 개수를 구하시오.

```python
import pandas as pd
df = pd.read_csv('data4-3.csv')
```

```python
df.head()
```

```python
df['date_to_date'] = pd.to_datetime(df['date_added'])
```

```python
df['date_to_date'] = df['date_to_date'].dt.to_period('M')
```

```python
cond31 = df['date_to_date'] == '2018-01'
```

```python
df[cond31]
```

```python
cond32 = df['country'] == 'United Kingdom'
```

```python
len(df[cond31&cond32])
```
