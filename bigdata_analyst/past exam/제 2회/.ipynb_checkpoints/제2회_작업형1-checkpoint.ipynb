# 문제2 --------------------------------------------------

주어진 데이터셋에서 앞에서부터 순서대로 80퍼센트 데이터만 활용해 f1 컬럼 결측치를 
중앙값으로 채우기 전후의 표본 표준편차를 구하고,두 표본 표준편차의 차이를 계산하시오. 
(두 표본 표준편차의 차이는 절댓값으로 계산하고,반올림하여 소수 둘째자리까지 계산)

```python
import pandas as pd
df = pd.read_csv('members.csv')
```

```python
df.head()
print(len(df))
```

```python
###################################################
line = int(len(df)*0.8)
print(line)
df= df.iloc[:line]

std1 = df['f1'].std()
print(std1)
med = df['f1'].median()
print(med)
df['f2'] = df['f1'].fillna(med)

std2 = df['f2'].std()
print(std2)
result = abs(std1-std2)
print(round(result,2))
###################################################
```

```python
# 앞에서부터 80% 데이터 활용
int(len(df)*0.8)
```

```python
df = df.iloc[:int(len(df)*0.8)]
```

```python
len(df)
```

```python
# f1 컬럼 결측치를 중앙값으로 채우기 전의 표본 표준편차
before_f1 = df['f1'].std()
print(before_f1)
```

```python
# f1 컬럼 결측치를 중앙값으로 채우기 후의 표본 표준편차
f1_median = df['f1'].median()
```

```python
print(f1_median)
```

```python
df['f1_fill'] = df['f1'].fillna(f1_median)
```

```python
after_f1 = df['f1_fill'].std()
```

```python
print(after_f1)
```

```python
round(abs(before_f1 - after_f1), 2)
```

# 앞으로 처음부터 한 번씩 다 쳐보기 또는 종료 후 함수 한꺼번에 모아 치기

주어진 데이터셋의 age 컬럼의 이상치를 모두 더하시오. 이상치는 평균으로부터
표준편차 * 1.5를 벗어나는 영역을 이상치라고 판단한다. 

```python
import pandas as pd
df = pd.read_csv('members.csv')
```

```python
df.head()
```

```python
# age의 이상치 구하기
age_mean = df['age'].mean()
print("age_mean:", age_mean)

baseline = df['age'].std() * 1.5
print("baseline:", baseline)

# 평균으로부터 표준편차 1.5 를 벗어난다는게 이상치라는게 무슨말일까. 한국말을 이해하지 못하겠다.
cond1 = df['age'] > (age_mean+ baseline)
cond2 = df['age'] < (age_mean-baseline)

print(df['age'][cond1|cond2].sum())
print(sum(df['age'][cond1|cond2]))
print(sum(df['age'][cond1]))

```
