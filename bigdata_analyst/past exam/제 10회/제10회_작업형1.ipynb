# 문제 1¶
제공된 데이터(subject_performance.csv)는 학생들이 각 문제를 풀었을 때의 정답 여부를 기록한 자료이다. 각 문제는 대주제와 소주제로 분류되어 있다. 아래 수행 순서에 따라 데이터를 처리한 뒤, 구한 값을 [제출 형식]에 맞춰 답안 페이지에 입력하시오.

제공 데이터

major_topic : 대주제
sub_topic : 소주제
student_id : 학생 ID
is_correct : 정답 여부(1: 정답, 0: 오답)
수행 순서

소주제별 정답률을 계산한다.
정답률 = (해당 소주제의 정답 개수) / (해당 소주제의 전체 문제 풀이 수)
소주제들의 정답률 중 3번째로 높은 값을 구하시오.
단, 정답률이 같은 경우 하나로 취급한다. 정답률이 0.7, 0.7, 0.6, 0.5인 경우, 세 번째로 높은 값은 0.5이다.
[제출형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/subject_performance.csv")
df.head()
```

```python
all_q_counts = len(df)
print(all_q_counts)

grouped = df.groupby('sub_topic')['is_correct'].agg(['count', 'sum'])
grouped['정답률'] = grouped['sum']/grouped['count']

#grouped['정답률'] = (grouped['is_correct'] / all_q_counts)
result = grouped['정답률'].sort_values(ascending=False).drop_duplicates()

#result.nlargest(3)
result = result.nlargest(3).values[-1]

round(result, 3)
```

# [!] 중복 제거 drop_duplicates

문제2¶
제공된 데이터(cafe_sales.csv)는 2024년 4월부터 2025년 12월까지의 카페 메뉴 판매 이력이다. 소문항을 각각 구한 다음 [제출 형식]에 맞춰 답안 페이지에 입력하시오.

제공 데이터

order_date : 주문 날짜 (예 2024.05.24 13:22)
category : 음료/디저트 종류 (예 coffee, tea …)
item : 메뉴명
price : 판매 금액
소문항

2-1. 연-월별 총 매출액을 계산하여 큰 순서대로 정렬했을 때, 2번째로 큰 매출액을 입력하시오.

2-2. 연-월별 총 매출액이 4번째로 큰 연-월을 찾으시오. 해당 연-월의 데이터에서 카테고리별 매출 합계를 계산하고, 그중 최댓값을 입력하시오.

[제출 형식] 정수로 작성

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/cafe_sales.csv")
df.head()
```

```python
# 2-1. 연-월별 총 매출액을 계산하여 큰 순서대로 정렬했을 때, 2번째로 큰 매출액을 입력하시오.
print(type(df['order_date'][0]))

df['연월'] = pd.to_datetime(df['order_date']).dt.to_period('M')

result = df.groupby('연월')['price'].sum().nlargest(2).values[-1]

print(result)
```

```python
#2-2. 연-월별 총 매출액이 4번째로 큰 연-월을 찾으시오. 해당 연-월의 데이터에서 카테고리별 매출 합계를 계산하고, 그중 최댓값을 입력하시오.

df.groupby('연월')['price'].sum().nlargest(4).index[-1] # Period('2024-10', 'M')

#2024-10
cond1 = df['연월'] == df.groupby('연월')['price'].sum().nlargest(4).index[-1]

result = df[cond1].groupby('category')['price'].sum().max()
print(result)

```

# 문제 3 -----------------------------------------------

문제 3
제공된 데이터(hamspam.csv)는 문자 메시지 내용과 스팸 여부를 담고 있다. 아래 수행 순서에 따라 데이터를 처리한 뒤, 구한 값을 [제출 형식]에 맞춰 답안 페이지에 입력하시오.

제공 데이터

label : 메시지 구분 (spam: 스팸, ham: 정상)
text : 메시지
수행 순서

각 메시지의 단어 개수를 띄어쓰기(공백) 기준으로 계산하시오.
스팸 메시지의 평균 단어 개수와 정상 메시지의 평균 단어 개수를 각각 구하시오.
두 평균의 차이의 절댓값을 구하여 입력하시오.
[제출형식] 반올림하여 소수 셋째 자리까지 작성

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/hamspam.csv")
df.head()
```

# df["단어수"] = df["단어"].str.len()

```python
# 각 메시지의 단어 개수를 띄어쓰기(공백) 기준으로 계산하시오.
#print(df['text'].str.split(' ')[0])
#print(df['text'].str.count(' '))
df['단어개수'] =  df['text'].str.count(' ')

#스팸 메시지의 평균 단어 개수와 정상 메시지의 평균 단어 개수를 각각 구하시오.
ham_c = df.groupby('label')['단어개수'].mean()['ham']
spam_c = df.groupby('label')['단어개수'].mean()['spam']

# 두 평균의 차이의 절댓값을 구하여 입력하시오
print(abs(ham_c - spam_c))

#반올림하여 소수 셋째자리

round(abs(ham_c - spam_c), 3)
```
