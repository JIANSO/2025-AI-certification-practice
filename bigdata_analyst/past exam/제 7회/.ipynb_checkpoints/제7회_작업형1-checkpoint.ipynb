# 문제 2.-----------------------------------------------------------------------------

DE1 ~ DE77 컬럼 중 주가지수의 종가 'close'와 가장 상관 관계가 높은 변수를 찾아,
해당 변수의 평균값을 구하시오(반올림하여 소수 넷째 자리까지 계산)

```python
import pandas as pd
df = pd.read_csv('stock_market.csv')
```

```python
df.head()
```

```python
df.corr()['close'].sort_values()
#DE12     0.055408
```

```python
# [!!] 상관관계는 abs() 절대값을 사용해 구한다.
df_corr = df.corr()['close'].abs()
```

```python
col = df_corr.loc['DE1':'DE77'].idxmax()
```

```python
# 그 값의 평균값
round(df[col].mean(), 4)
```

# [!] 첫 시작이 DE로 시작해야 해

```python
round(0.055408, 4)
```

# 문제 3.-----------------------------------------------------------------------------

IQR을 이용해 이산화탄소 이상치 수를 찾으시오.
# 자료가 없어 8과의 chem을 빌림. 

```python
import pandas as pd
df = pd.read_csv('chem.csv')
```

```python
df.head()
```

```python
df_co = df['co']
```

```python
q1= df_co.quantile(0.25)
q3= df_co.quantile(0.75)
iqr = q3-q1
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr
```

```python
outliers = df[ (df_co<=lower) & (df_co >=upper)]
len(outliers)
```
