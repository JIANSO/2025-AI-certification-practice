# 문제
W-mart 매장의 판매 데이터를 기반으로 판매액을 예측하시오.

```python
import pandas as pd
train= pd.read_csv('mart_train.csv')
test= pd.read_csv('mart_test.csv')
```

```python
train.info()
target = train.pop('total')
```

```python
test.info()
```

```python
# object가 너무 많아서 Label Encoder 사용해야함
```

```python
train.isna().sum()
```

```python
test.isna().sum()
```

```python
train.shape, test.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

# [!] concat 방법 바로 떠올릴 수 있도록

```python
concat = pd.concat([train, test])
```

```python
concat.shape
```

```python
concat.head()
#삭제할 건 없어보인다.
```

# [!] y should be a 1d array, got an array of shape (1000, 9) instead.

```python
obj_col = concat.select_dtypes(include="O").columns # object 컬럼만 추출
obj_col #컬럼만 리스트
```

```python
for col in obj_col : 
    le = LabelEncoder()
    concat[col] = le.fit_transform(concat[col])
```

```python
concat.head()
```

```python
#다시 나누기
train = concat.iloc[:700,]
test = concat.iloc[700:,]
```

```python
train.shape, test.shape
```

```python
# 트레인 세트 분기하기
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
model = rf.fit(X_train, y_train)
```

```python
pred = model.predict(X_val)
```
