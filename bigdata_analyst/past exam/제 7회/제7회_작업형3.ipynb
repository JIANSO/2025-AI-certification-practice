# 문제1----------------------------------------------------------------

```python
import pandas as pd
df = pd.read_csv("clam.csv")
```

```python
train = df.iloc[:210]
test = df.iloc[210:]
```

##### 문제 1-1, weight를 독립변수, gender를 종속변수로 사용해 로지스틱 회귀 모델을 만들고, weight 변수가 무게가 한 단위 증가할 때 수컷일 오즈비 값은?(반올림하여 소수 넷째자리)

```python
from statsmodels.formula.api import logit
```

```python
formular = 'gender~weight'
model = logit(formular, data=train).fit()
```

```python
model.summary()
```

```python
import numpy as np
round(np.exp(model.params['weight']), 4)
```

##### 문제 1-2, gender를 종속변수로 하고 나머지 변수들(age, length, diameter, height, weight)을 독립변수로 사용하는 로지스틱 회귀 모델을 적합시킨 후, 잔차 이탈도를 계산하시오(반올림 소수 둘째자리)

```python
model2 = logit('gender~age+length+diameter+height+weight', data=train).fit()
```

```python
model2.llf # 이름 출력이 안 되는 중 -> 로그 우도
```

```python
round((model2.llf*2),2)
```

# [!] 잔차이탈도는 로그 우도에 2가 아니라 -2를 곱한것이다.

```python
round((model2.llf*-2),2)
```

##### 문제 1-3, 독립변수 weight만 사용해 학습한 모델에서 test 데이터의 gender를 예측하고, error rate(오류율)을 구하시오. (반올림 소수 셋째자리)

```python
# 어떻게 독립변수 weight만 사용하지?
formular = 'gender~weight'
model3 = logit(formular, data=train).fit()
```

```python
# [!] 로지스틱 회귀에서 predict할 때 test에서 target 꼭 빼기
target = test.pop('gender')
```

```python
result = model3.predict(test)
```

```python
pred = (result>0.5).astype(int)
```

# [?] 오류율? -> 1-accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(target, pred)
```

```python
error_rate = 1 - accuracy
```

```python
print(round(error_rate, 3))
```
