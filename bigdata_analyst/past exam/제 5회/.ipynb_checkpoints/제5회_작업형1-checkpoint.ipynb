# 문제1------------------------------------------------------------------------------------

종량제 봉투 종류가 규격봉투고 종량제 봉투 용도가 음식물쓰레기인 2L 가격 평균을 구하시오.
단 가격이 0이면 평균에서 제외하고, 반올림 후 정수를 출력하시오.

```python
import pandas as pd
df = pd.read_csv('data5-1.csv')
```

```python
df.head()
```

```python
cond1 = df['종량제봉투종류'] == '규격봉투'
cond2 = df['종량제봉투용도'] == '음식물쓰레기'
```

```python
df[cond1 & cond2].head()
```

```python
#result = df[cond1 & cond2]['2ℓ가격'].mean()
# 가격이 0인것들은 평균에서 제외하라.
edited = df[cond1 & cond2]
```

```python
cond3 = edited['2ℓ가격'] != 0
```

```python
edited[cond3].head()
```

```python
result = edited[cond3]['2ℓ가격'].mean()
```

```python
result = round(result) #반올림 후 정수 출력이요 ㅠㅠ 이러면 틀립니다.
```

```python
print(result)
```

# 문제3------------------------------------------------------------------------------------

순 전입학생이 가장 많은 학교의 전체 학생 수를 구하시오
(순 전입학생==전입학생-전출학생)

```python
import pandas as pd
df = pd.read_csv("data5-3.csv")
```

```python
df.head()
```

```python
df['순전입학생'] = df['전입학생수(계)'] - df['전출학생수(계)']
```

```python
len(df['학교명'].unique())
```

```python
df.shape
```

```python
grouped = df.groupby(['학교명']).sum()
```

```python
best = grouped['순전입학생'].sort_values(ascending=False).idxmax()
```

```python
cond3 = df['학교명'] == best
```

```python
df[cond3]['전체학생수(계)']
```
