# 문제 -----------------------------------------------
차량의 특징 기반 데이터를 활용해 test 데이터의 중고차 가격을 예측하시오.

```python
import pandas as pd
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.shape, test.shape
```

```python
train.head()
```

```python
target = train.pop("price")
```

```python
train.isna().sum(), test.isna().sum()
```

```python
train["is_train"] = 1
test["is_train"] = 0

concat_data = pd.concat([train, test])
```

```python
concat_data = pd.get_dummies(concat_data)
```

```python
concat_data.shape
```

```python
concat_data.head()
```

```python
train = concat_data[ concat_data['is_train'] == 1].drop("is_train", axis=1)
test = concat_data[ concat_data['is_train'] == 0].drop("is_train", axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train, target, test_size =0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
```

```python
model = rf.fit(X_train, y_train)
```

```python
model_pred = model.predict(X_val)
```

```python
print(model_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
result = mean_squared_error(y_val, model_pred, squared=True)  # 또 바로 생각안난다
```

```python
result
```

```python
final_result = model.predict(test)
```

# [!] 내가 model.predict(test) 의 결과값을 넘긴다는 것을 잊지말기.
