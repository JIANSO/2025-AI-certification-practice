건물 데이터를 기반으로 건물의 난방 부하(에너지 효율성)을 예측하시오. 

```python
import pandas as pd
train = pd.read_csv("energy_train.csv")
test = pd.read_csv("energy_test.csv")
```

# [?] target 도 get_dummies

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
target = le.fit_transform(target)
```

```python
print(target)
```

```python
train.info()
```

```python
test.info()
```

```python
print(train.isna().sum(), test.isna().sum())
```

```python
#train_object = train.select_dtypes(include='O')
```

```python
train['is_train'] = 1
test['is_train'] = 0
concat_data = pd.concat([train, test])
```

```python
concat_data = pd.get_dummies(concat_data)
```

```python
concat_data.head()
```

```python
train = concat_data[ concat_data['is_train'] == 1].drop('is_train', axis=1)
test = concat_data[ concat_data['is_train'] == 0].drop('is_train', axis=1)
```

```python
train.head()
```

```python
test.head()
```

# [!] X, y 순서 헷갈리지 말기

```python
# train, test split
from sklearn.model_selection import train_test_split

# X, y 순서 헷갈리지 말기
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
model = rf.fit(X_train, y_train)
```

```python
pred = model.predict(X_val)
```

```python
from sklearn.metrics import f1_score
```
