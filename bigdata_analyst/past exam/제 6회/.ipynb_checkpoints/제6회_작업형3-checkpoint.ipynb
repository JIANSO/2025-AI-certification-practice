# 문제 1 -----------------------------------------------------------------------------------
감기약 위약을 투여받은 환자의 부작용은, 항암약 위약을 투여받은 환자의 부작용과 차이가 있는가?

##### 문제 1-1 
항암약을 투여받은 환자 중 '이상없음'의 비율을 0과 1사이로 구하시오

```python
# 항암약 부작용
import pandas as pd
df = pd.DataFrame({ 
    "항암약" : [4,4,3,4,1,4,1,4,1,4,4,2,1,4,2,3,2,4,4,4]
})
```

```python
print(df.head(10))
```

# [!] value_counts에 normalize=True 옵션을 설정하면, 각 값의 상대적 비율로 반환된다.

```python
#답
df_counts = df.value_counts()
print(df_counts)
df_len = len(df)
print(df_len)
```

```python
# 답
result = (df_counts.values[0]/df_len)
print(result)
```

##### 문제 1-2
감기약의 예상 부작용 비율과 항암약의 부작용 관찰값이 통계적으로 유의미하게 차이가 있는지 확인하려 한다.
카이제곱 검정을 사용해 검정 통계량을 구하시오.

```python
#부작용 비율이라고 하면. 따로 넣어줘야 하나?
```

```python
# 감기약의 부작용 비율 
```

```python
a = [0.1, 0.05, 0.15, 0.7]
b = [4/20, 3/20, 2/20, 11/20]
```

```python
#두 집단의 분포? 확인 -> 적합도?
from scipy import stats
result = stats.chi2_contingency(a,b)
```

```python
from scipy.stats import chisquare

expected_value = a # 여기에 각 20을 곱해줘야 한다.
expected_value = [x * 20 for x in expected_value]
observed_value = df.value_counts().sort_index().to_list()

result = chisquare(f_obs = observed_value, f_exp =expected_value)
print(result)
```

##### 문제 1-3
위의 p-value를 구하시오. => 0.07266054733847573

# 문제 2 -----------------------------------------------------------------------------------
다중 선형 회귀 모델을 구축하고, 각 소문제에 답하시오.

```python
import pandas as pd
df = pd.read_csv("data6-3-2.csv")
print(df.head(3))
```

##### 문제 2-1 
주어진 데이터를 기반으로 다중 선형 회귀 모델을 구축하고, 독립변수 o3의 회귀 계수를 구하시오

```python
from statsmodels.formula.api import ols
model = ols('temperature~solar+wind+o3', data=df).fit()
```

```python
model.params['o3']
```

##### 문제 2-2 
문제 2-1에서 적합한 모델에서 solar와 o3이 고정된 상태에서 wind의 세기가 증가함에 따라 temperature가 감소한다.
이 회귀 모델에서 wind의 회귀 계수에 대한 p-value를 구하시오.

```python
model.pvalues['wind']
```

문제 2-1에서 적합한 모델에서 solar:100, wind:5, o3:30일 때, 예측값을 구하시오.
