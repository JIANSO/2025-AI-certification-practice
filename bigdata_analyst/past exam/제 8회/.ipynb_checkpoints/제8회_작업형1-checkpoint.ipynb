# 문제 3------------------------------------------------------------------------------------

##### 문제1) 주어진 데이터에서 'co'와 'nmhc' 컬럼을 각각 Min-max 스케일링 하시오


```python
import pandas as pd
df = pd.read_csv('chem.csv')
```

```python
df.head()
```

```python
co_min_X = df['co'].min()
co_max_X = df['co'].max()
```

```python
df['co_scaled'] = (df['co'] - co_min_X ) / (co_max_X - co_min_X)
```

```python
nmhc_min_X = df['nmhc'].min()
nmhc_max_X = df['nmhc'].max()
```

```python
df['nmhc_scaled'] = (df['nmhc'] - nmhc_min_X ) / (nmhc_max_X - nmhc_min_X)
```
