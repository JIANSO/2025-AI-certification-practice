### 문제. 통신사에서 고객에게 청구될 총 금액을 예측하시오

```python
import pandas as pd
train = pd.read_csv("churn_train.csv")
test = pd.read_csv("churn_test.csv")
```

```python
train.shape, test.shape
```

```python
train.head()
```

# [!] customerID는 그냥 버리는 컬럼이야. 모델 입장에선 완전 노이즈라서, X에서 빼는 게 정답이다. 버리는 컬럼을 볼 줄 알아야 한다.

```python
target = train['TotalCharges']
```

```python
target.shape
```

```python
X = train.drop('TotalCharges', axis=1)
```

```python
X.pop('customerID')
test.pop('customerID')
```

```python
X.isna().sum()
```

```python
test.isna().sum()
```

```python
X.info()
```

# [!] object 를 확인하고, 항상 변환해주기 - get_dummies 또는 labelEncoder(숙지 필수)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

# [!] cols = X.select_dtypes(include='0').columns

```python
cols = X.select_dtypes(include='O').columns
```

```python
for col in cols  :
    X[col] = le.fit_transform(X[col])
    test[col] = le.transform(test[col])
```

```python
#X_encoded = pd.get_dummies(X)
```

```python
#X_encoded.shape
```

```python
X.head()
```

```python
test.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
model = rf.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
result = mean_absolute_error(y_test, pred)
```

# [!] 제출할 pred는 최종 모델에 test 돌린 것. test도 잘 처리해줘야 함.
