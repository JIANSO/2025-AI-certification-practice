```python
import pandas as pd
df = pd.read_csv('churn.csv')
```

##### 문제 1-1. 주어진 데이터에서 로지스틱 회귀 분석을 수행해, 유의확률이 0.05 이상인 유의하지 않은 독립변수의 개수를 구하시오

```python
from statsmodels.formula.api import logit
```

```python
df.head()
```

```python
model = logit('Churn~AccountWeeks+ContractRenewal+DataPlan+DataUsage+CustServCalls+DayMins+DayCalls+MonthlyCharge+OverageFee+RoamMins', data = df).fit()
```

```python
#답 :
sum(model.pvalues >= 0.05)
```

# 문제2 ----------------------------------------------------------------------------

```python
import pandas as pd
df = pd.read_csv('piq.csv')
```

##### 문제 2-1. 주어진 데이터를 이용해 종속변수와 독립변수로 다중 선형 회귀 분석을 수행하시오. 이때 유의 확률이 가장 작은 변수의 회귀 계수 값을 구하시오(반올림 소수 셋째자리)

```python
df.head()
```

```python
from statsmodels.formula.api import ols
```

```python
formula = 'PIQ~Brain+Height+Weight'
model = ols(formula, data=df).fit()
```

```python
model.summary()
```

```python
# 답 : 
#model.pvalues['Brain']
#round(model.pvalues.sort_values().iloc[0])
model.pvalues.sort_values().iloc[0]
```

# [!] 회귀 계수를 구하는지, p-value를 구하는지 헷갈리지 말기

```python
# 그 값의 회귀 계수 값이니까 -> 
round(model.params['Brain'],3)
```

##### 문제 2-2. 문제 2-1에 적합한 모델의 결정 계수 값을 구하시오(반올림하여 소수 둘째 자리까지)

# [?] 결정계수? summray 하면 볼 수 있음. rsquared
