# 문제 1 -----------------------------------------------------------

##### 문제 1-1) 
1) 총 대출액은 신용대출과 담보대출의 합이다.
2) 각 지역코드 내에서, 성별별로 '총대출액'의 합계를 계산하시오.
3) 성별 간 총대출액의 차이가 가장 큰 지역코드를 구하시오.

```python
import pandas as pd
df = pd.read_csv('loan.csv')
```

```python
df.head()
```

```python
df['총대출액'] = df['신용대출'] +  df['담보대출']
```

```python
# 지역코드 내 # 성별별로 총 대출액 합계
grouped = df.groupby(['지역코드', '성별'])['총대출액'].sum()
```

```python
#성별 간 총 대출액의 차이 구하기
unstack_grouped = grouped.unstack(level=-1)
unstack_grouped['성별간차이'] = abs(unstack_grouped[1]-unstack_grouped[2])
result = unstack_grouped.sort_values(by='성별간차이', ascending=False).index[0]
```

```python
print(result)
```

##### 2) 근속연수의 결측치는, 같은 부서 내에서 동일한 성과등급을 가진 직원들의 평균 근속연수로 대체. 이 때 평균 근속연수는 소수점 이하를 절사하여 정수로 변환하여 사용한다.

-> transform 사용방법 및 적용 방법 다시 확인하기

```python
df.groupby(['부서', '성과등급']).agg('mean')
```

```python
# 같은 부서 내에서 동일한 성과등급을 가진 직원들의 평균 근속연수
grouped = df.groupby(['부서', '성과등급'])
#grouped['평균근속연수']  = grouped['근속연수'].transform('mean')
#근속연수의 결측치 
df['평균근속연수'] = round(grouped['근속연수'].transform('mean'),0).astype(int)
#df.groupby(['부서', '성과등급']).mean()

df['근속연수'].fillna(df['평균근속연수'], inplace=True)
```
