```python
import pandas as pd
df = pd.read_csv('loan.csv')
df
```

```python
df.groupby('지역코드')['신용대출'].agg(['mean', 'max'])
```

```python
df.groupby('지역코드')['신용대출'].transform(lambda x: '출력')
```

```python
df['지역코드'].isin([4100000000]).value_counts()
```

```python
df['지역코드'].between(4100000000, 4100000001, inclusive="right")
```
