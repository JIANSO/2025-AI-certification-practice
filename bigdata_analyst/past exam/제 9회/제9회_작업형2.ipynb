### 농작물에서 농약 검출 여부를 예측하시오

- 제공된 목록 : farm_train.csv, farm_test.csv
- 예측할 컬럼 : 농약검출여부(0:미검출, 1:검출, 2:재검사필요)

제출한 모델의 성능은 macro F1 Score로 평가

```python
import pandas as pd
train = pd.read_csv('farm_train.csv')
test = pd.read_csv('farm_test.csv')
```

```python
test.info()
```

#### 개인문제풀이---------------------------------------------------------------

```python
from sklearn.model_selection import train_test_split
```

```python
train_copy = train
test_copy = test
```

```python
train_copy.isna().sum()
```

```python
test_copy.isna().sum()
```

```python
train_copy.head()
```

```python
train_copy.loc[:,'등급']
```

```python
concat_copy = pd.concat([train_copy, test_copy])
```

```python
concat_copy
```

```python
#라벨 인코딩 # 다시 학습 필요
from sklearn.preprocessing  import LabelEncoder

o_list = ['지역', '작물종류', '토양유형', '등급']

for col in o_list : 
    le = LabelEncoder()
    concat_copy[col] = le.fit_transform(concat_copy[col])
```

```python
train_copy = concat_copy[:1000:]
```

```python
concat_copy.shape
```

```python
train_copy = concat_copy.iloc[:4000,:]
```

```python
train_copy.shape
```

```python
test_copy = concat_copy.iloc[4000:, :]
```

```python
test_copy.shape
```

```python
y = train_copy.pop('농약검출여부')
```

```python
x = train_copy
```

```python
x.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf =  RandomForestClassifier()
```

```python
model = rf.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score
```

```python
###???
result = f1_score(y_test, pred, average='macro')
```

```python
result
```

```python
test_copy
```

```python
test = test_copy.drop('농약검출여부', axis=1)
```

```python
result_pred = model.predict(test)
```

```python
result_pred
```

```python
pd.DataFrame({'pred':result_pred}).to_csv("result.csv", index=False)
```
