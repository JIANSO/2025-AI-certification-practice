# 문제1. 다중 선형 회귀

##### 문제 1-1 
train 데이터에서 c1, c2, c3, c4, c5를 독립변수로, design을 종속변수로 하는 다중 회귀 모형을 적합한 후,
절편을 제외한 독립변수 중 p-value가 0.05보다 작은 변수의 개수를 구하시오.

```python
import pandas as pd
df = pd.read_csv('design.csv')
```

```python
df.head()
```

##### [!] statsmodels OLS의 predict()는 y를 요구하지 않는다

```python
##### 여기서 train, test를 나눈다
cond1 = df['id'] <= 140
cond2 = df['id'] > 140
train = df[cond1].copy()
test = df[cond2].copy()
```

```python
#다중 회귀 모델
from statsmodels.formula.api import ols
model = ols('design ~ c1+c2+c3+c4+c5', data=train).fit()
```

```python
model.summary()
```

```python
# 문제 1=1 답 : 3개(c1, c2, c4)
```

# [?] 상수항은 제외하고, p-value가 0.05보다 작은 독립 변수의 계수를 계산하는 식

### 상수항을 제외하기 때문에 pvalues[1:] 을 한다

```python
print(model.pvalues[1:] <0.05)
```

```python
sum(model.pvalues[1:] <0.05)
```
