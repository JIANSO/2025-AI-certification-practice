### 문제 2 -------------------------------------------------------------------------------------

1) 검거율은 검거건수 / 발생건수이다.
2) 각 연도에서 가장 높은 검거율을 기록한 범죄유형을 찾으시오.
3) 해당 범죄유형의 검거건수를 구한 후 이 값들을 모두 합산하시오(정수로 구하시오).

```python
import pandas as pd
df = pd.read_csv('crime.csv')
```

```python
df
```

```python
df_gumgu = df[df['구분'] == '검거건수']
#어떤식으로 컬럼을 왔다갔다하지?
test = pd.pivot_table(columns='연도', data =df_gumgu)
test
```

```python
result = 0
year_list = [2014, 2015, 2016, 2017, 2018, 2019, 2020]
```

```python
for year in  year_list:
    result += test[year].max()
```

```python
result
```

#### 해답

```python
# 해답 - 발생건수와 검거건수 분리
cond1 = df['구분'] == '발생건수'
cond2 = df['구분'] == '검거건수'
df1  = df[cond1].iloc[:,2:]
df2  = df[cond2].iloc[:,2:]
```

```python
# 해답 - 검거율 계산
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
df3 = df2/df1

```

```python
df3
```

```python
listbox = df3.idxmax(axis=1)
'''
각 행(row)을 기준으로 보면서 가장 큰 값이 어느 column에 있는지 그 column 이름을 돌려줍니다.
'''
listbox
```

```python
# 해답
original = 0
for index, item in enumerate(listbox) :
    original += df2.loc[index, item]
    
original
```

#### 2) 근속 연수의 결측치는, 같은 부서 내에서 동일한 성과등급을 가진 직원들의 평균 근속연수로 대체한다. 이때 평균 근속연수는 소수점 이하를 절사하여 정수로 변환하여 사용한다.

#해답---------------------------------------------------------------------------

```python
gm = df.groupby(['부서', '성과등급'])['근속연수'].transform("mean")
```

```python
gm = gm.astype(int)
```

```python
df['근속연수'] = df['근속연수'].fillna(gm)
```

#해답---------------------------------------------------------------------------

```python
# 평균 근속연수 찾기
grouped = df.groupby(['부서', '성과등급']).mean(numeric_only=True)
```

```python
# 그룹 바이의 결과를 어떻게 대체하지? 
grouped['평균근속연수'] = grouped['근속연수'].astype(int)
```

```python
grouped.loc[('Finance', 'A')]['평균근속연수']
```

```python
# groupe된 두 인덱스에 접근하기 위해서는? ㅠㅠ
부서 = grouped.index.get_level_values('부서').unique()
성과등급 = grouped.index.get_level_values('성과등급').unique()
```

```python
for i in list(부서) :
    for j in list(성과등급) :
        df_copy = df[ (df['부서']==i)&(df['성과등급']==j)].fillna(int(grouped.loc[(i, j)]['평균근속연수']))
       
```

#### 3) 각 직원의 '연봉/근속연수' 값을 계산하여, 그 값이 세 번째로 큰 직원의 근속 연수를 구하시오

```python
df_copy['연봉/근속연수'] = (df_copy['연봉']/df_copy['근속연수'])
```

```python
df_year = df_copy.nlargest(3, '연봉/근속연수')
A = df_year.iloc[-1]['근속연수']
```

```python
df_copy.sort_values(by='연봉/근속연수', ascending=False)
# 세 번째로 큰 직원의 근속 연수는 -> 3
```

#### 4) 각 직원의 '연봉/만족도' 값을 계산하여, 그 값이 두 번째로 큰 직원의 교육 참가 횟수를 구하시오

```python
df_copy['연봉/만족도'] = (df_copy['연봉']/df_copy['만족도'])
```

```python
df_copy.sort_values(by='연봉/만족도', ascending=False)
# 두 번째로 큰 직원의 교육참가횟수 -> 5
```
